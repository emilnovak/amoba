```python
def reset():
    clear()
    print("Restart the kernel to continue playing!")
    
    while(1):
        ;
```

```python
bounds_x = 100
bounds_y = 25

term_char = '#'
empty_char = ' '
player_char = 'X'
ai_char = 'O'

rows = []
winningTiles = []

winState = False
first_move = True


def start(width, height):
    
    global rows
    global winningTiles
    global player_pos
    global first_move
    global bounds_x
    global bounds_y
    
    bounds_x = width
    bounds_y = height
    
    first_move = True
    
    player_pos = (int(round(bounds_y / 2, 0)), int(round(bounds_x / 2, 0)))


    for i in range(0, bounds_y):
        elements = []
        for j in range(0, bounds_x):
            if i == 0 or i == bounds_y-1 or j == 0 or j == bounds_x-1:
                elements.append(term_char)
            else:
                elements.append(empty_char)
        rows.append(elements)

    (y, x) = player_pos

    # rows[y][x] = player_char
    clear()
    render()
```

```python
def validTile(pos):
    
    (y, x) = pos
    
    if rows[y][x] != empty_char or y == 0 or x == 0 or y == bounds_y - 1 or x == bounds_x - 1 or winState:
        return False
    
    if first_move:
        return True

    if rows[y - 1][x] == player_char or rows[y - 1][x] == ai_char \
      or rows[y + 1][x] == player_char or rows[y + 1][x] == ai_char \
      or rows[y][x - 1] == player_char or rows[y][x - 1] == ai_char \
      or rows[y][x + 1] == player_char or rows[y][x + 1] == ai_char \
      or rows[y - 1][x - 1] == player_char or rows[y - 1][x - 1] == ai_char \
      or rows[y + 1][x - 1] == player_char or rows[y + 1][x - 1] == ai_char \
      or rows[y - 1][x + 1] == player_char or rows[y - 1][x + 1] == ai_char \
      or rows[y + 1][x + 1] == player_char or rows[y + 1][x + 1] == ai_char:
        
        return True
    
    return False
        
```

```python
def validNext(currentTile, possibilitiles, direction):
    # direction: 0 - down, 1 - down-right, 2 - right, 3 up-right
    
    (curr_y, curr_x) = currentTile
    
    if direction == 0:
        for tile in possibilitiles:
            (p_y, p_x) = tile
            if(p_y == curr_y + 1 and p_x == curr_x):
                return (p_y, p_x)
            
    elif direction == 1:
        for tile in possibilitiles:
            (p_y, p_x) = tile
            if(p_y == curr_y + 1 and p_x == curr_x + 1):
                return (p_y, p_x)
            
    elif direction == 2:
        for tile in possibilitiles:
            (p_y, p_x) = tile
            if(p_y == curr_y and p_x == curr_x + 1):
                return (p_y, p_x)
            
    elif direction == 3:
        for tile in possibilitiles:
            (p_y, p_x) = tile
            if(p_y == curr_y - 1 and p_x == curr_x + 1):
                return (p_y, p_x)
            
    return False


def checkForWinner(char):
    global winningTiles
    global winState
    
    occuppiedTiles = []
    for i, row in enumerate(rows):
        for j, tile in enumerate(row):
            if tile == char:
                occuppiedTiles.append((i, j))
                
    # print('occuppiedTiles: ', occuppiedTiles)
    
    for tile in occuppiedTiles:
        checkList = occuppiedTiles.copy()
        checkList.remove(tile)
        # print(checkList)
        
        for direction in range(0, 4):
            tileStreak = []
            tileStreak.append(tile)
        
            tmpTile = True
            while tmpTile != False:
                tmpTile = validNext(tile, checkList, direction)
                if(tmpTile != False):
                    tileStreak.append(tmpTile)
                    tile = tmpTile
            # print('tileStreak: ', tileStreak)
            # print('direction: ', direction)
            if len(tileStreak) >= 5:
                # TODO winning
                print(char, "has won! Please reset the Game")
                winState = True
                winningTiles = tileStreak
                return True
    return False
    
```

```python
def findPattern(pattern, anchor_pos):
    
    a_i, a_j = anchor_pos
    
    empty_spaces = 0
    occupied_tiles = 0
    offensive_tiles = 0
    
    
    for i, row in enumerate(pattern):
        for j, tile in enumerate(row):
            if a_i + i > bounds_y or a_j + j > bounds_x:
                return
            if tile == 'B':
                offensive_tiles += 1
                if rows[i + a_i][j + a_j] == ai_char:
                    offensive_tiles -= 1
            if tile == '#' or tile == 'O':
                empty_spaces += 1
                if rows[i + a_i][j + a_j] == empty_char:
                    empty_spaces -= 1
            if tile == 'X':
                occupied_tiles += 1
                if rows[i + a_i][j + a_j] == player_char:
                    occupied_tiles -= 1
    if empty_spaces == 0 and occupied_tiles == 0 and offensive_tiles == 0:
        return True
    return False
            
        
```

```python

def aiTick():
    global rows
    
    validMoves = []
    
    for i, row in enumerate(rows):
        for j, tile in enumerate(row):
            if validTile((i, j)):
                validMoves.append((i, j))
    #print(validMoves)
    
    probableMoves = []
    
    for tile in validMoves:
        for index, pattern in enumerate(patterns):
                    
            if findPattern(pattern, tile):
                # print("found pattern", pattern)
                for i, row in enumerate(pattern):
                    for j, character in enumerate(row):
                        if pattern[i][j] == 'O':
                            (a_i, a_j) = tile 
                            probableMoves.append((i + a_i, j + a_j))
                            #(u, v) = random.choice(probableMoves)
                            (u, v) = probableMoves[0]
                            # print(pattern)
                            rows[u][v] = ai_char
                            checkForWinner(ai_char)
                            return
                            
            
                    
    if len(probableMoves) == 0:
        (u, v) = random.choice(validMoves)
    
    rows[u][v] = ai_char
    
    checkForWinner(ai_char)
```

```python
def gameLogic(action):
    
    global rows
    global player_pos
    global first_move
    (y, x) = player_pos
    
    # Move to render on release
    clear() 
    
    # only debug
    # print('The action is:', action)
    
    checkForWinner(player_char)
    
    # TODO: Add crude state machine
    
    if action == 'enter':
        # print(action)
        
        if(validTile(player_pos)):
            rows[y][x] = player_char
            
            if checkForWinner(player_char):
                render()
                return
            first_move = False
            aiTick()
        
    elif action == 'back':
        # print(action)
        reset()
        
    elif action == 'left':
        # print(action)
        
        if rows[y][x - 1] != term_char:
            player_pos = (y, x - 1)
        
    elif action == 'down':
        # print(action)
        
        if rows[y + 1][x] != term_char:
            player_pos = (y + 1, x)
        
    elif action == 'up':
        # print(action)
        
        if rows[y - 1][x] != term_char:
            player_pos = (y - 1, x)
            
    elif action == 'right':
        # print(action)
        
        if rows[y][x + 1] != term_char:
            player_pos = (y, x + 1)
            
    else:
        print('Invalid action!')
        
    # print("Is ValidTile: ", validTile(player_pos))
    
    render()
```

```python
def callback_enter(b):
    gameLogic('enter')
def callback_back(b):
    gameLogic('back')
def callback_left(b):
    gameLogic('left')
def callback_down(b):
    gameLogic('down')
def callback_up(b):
    gameLogic('up')
def callback_right(b):
    gameLogic('right')
def callback_reset(b):
    reset()

button_enter = widgets.Button(description='enter', icon='check')
button_back = widgets.Button(description='surrender', icon='legal')
button_left = widgets.Button(description='left')
button_down = widgets.Button(description='down')
button_up = widgets.Button(description='up')
button_right = widgets.Button(description='right')
button_reset = widgets.Button(description='reset')

button_enter.on_click(callback_enter)
button_back.on_click(callback_back)
button_left.on_click(callback_left)
button_down.on_click(callback_down)
button_up.on_click(callback_up)
button_right.on_click(callback_right)
button_reset.on_click(callback_reset)

def clear():
    clear_output()

def render():
    for i, row in enumerate(rows):
        for j, tile in enumerate(row):
            
            if winState:
                print(tile, end = '')
                for t in winningTiles:
                    (w_y, w_x) = t
                    if (i, j) == (w_y, w_x):
                        print('\b', end = '')
                        print(colored(tile, 'blue'), end='')
                        
            else:
            
                if player_pos == (i, j):
                    if rows[i][j] == empty_char:
                        if (validTile((i, j))):
                            if winState:
                                print(tile, end = '')
                            else:
                                print(colored(player_char, 'green'), end = '')
                        else:
                            print(colored(player_char, 'red'), end = '')
                    else:
                        print(colored(tile, 'red'), end = '')
                else:
                    print(tile, end = '')   
        print('')
    if winState:
        display(button_reset)
    else:
        display(widgets.HBox((button_enter, button_back, button_left, button_down, button_up, button_right)))
```
